**<h1>Data Cleaning</h1>**

<h2>Punctuation Removing</h2>

```python
import re
import string
```

```python
def cleanSt(text):
  punctuation_remove = ''.join(p for p in text if p not in string.punctuation)
  return punctuation_remove
```

```python
#punctuation removal
def cleanRe(text):
  punctuation_remove = re.sub(r'[^\w\s]' , '', text)
  return punctuation_remove
```

```python
Df['StringCleaning']=Df['description'].apply(cleanSt)
Df['RegexCleaning'] = Df['description'].apply(cleanRe)
```

<h2>Stop Words Removing </h2>

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwordsList = stopwords.words('english')
```

```python
count=0
item_to_remove = 'not'
for i in stopwordsList:
  count=count+1
  if(i == item_to_remove):
    print('found at index: ',count ,'in the original stopwords list')

#creating new stopwords list without "not"
new_stopwords_list = [x for x in stopwordsList if(x!= item_to_remove)]

cout = 0
for i in new_stopwords_list:
  cout=cout+1
  if(i == item_to_remove):
    print('found at index: ',cout)
```

```python
def cleanSt(text):
  stopwords_remove = [p for p in text.split() if(p not in new_stopwords_list)]
  stopwords_remove = ' '.join(stopwords_remove)
  return stopwords_remove
```

```python
Df['noSTop']=Df['description'].apply(cleanSt)
Df.iloc[:3, :]
```

<h2>Removing whitespaces</h2>

```python
#regex
def cleanRe(text):
  whitespace_remove = re.sub(r' +' , ' ',text) #more than one whitespaces removed
  whitespace_remove=whitespace_remove.strip() #for leading and trailing whitespaces
  return whitespace_remove
```

```python
Df['noWhitespace'] = Df['description'].apply(cleanRe)
```
